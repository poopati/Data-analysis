```python
#5
plt.figure(figsize=(10,5))
plt.hist(var_A['FSO'])
plt.xlabel("FSO_varA")
plt.ylabel("frequency")
plt.xticks(size=15)
plt.yticks(size=15)
plt.show(True)
```

```python
plt.figure(figsize=(10,5))
plt.hist(var_B['FSO'])
plt.xlabel("FSO_varB")
plt.ylabel("frequency")
plt.xticks(size=15)
plt.yticks(size=15)
plt.show(True)
```

```python
stats.ttest_ind(var_A['FSO'],var_B['FSO'])
```

### p<0.05,so the distribution of FSO for variant A and variant B are different.

```python
plt.figure(figsize=(10,5))
plt.hist(var_A['AV'])
plt.xlabel("AV_varA")
plt.ylabel("frequency")
plt.xticks(size=15)
plt.yticks(size=15)
plt.show(True)
```

```python
stats.ttest_ind(var_A['AV'],var_B['AV'])
```

### p<0.05,so the distribution of AV for variant A and variant B are different.

### 6

### Null hypothesis:PH values of the variants are not significantly different from one another 
### Alternative hypothesis:PH values of variants are significantly different from one another


```python
stats.ttest_ind(var_A['PH'],var_B['PH'])
```

### pvalue<0.05 hence we reject the null hypothesis
### therefore we can say that PH values of variants are significantly different from one another

```python
plt.figure(figsize=(10,5))
plt.xticks(size=15)
plt.yticks(size=15)
plt.hist(var_A['Q'])
plt.xlabel("Qa")
plt.ylabel("frequency")
plt.title("Frequency of Q")
plt.show(True)
```

```python
plt.figure(figsize=(10,5))
plt.xticks(size=15)
plt.yticks(size=15)
plt.hist(var_B['Q'])
plt.xlabel("Qb")
plt.ylabel("frequency")
plt.title("Frequency of Q")
plt.show(True)
```

### Null hypothesis:Q values of both the variants does not vary
### Alternative hypothesis:Q values of both the variants vary

```python
stats.ttest_ind(var_A['Q'],var_B['Q'])
```

### pvalue<0.05 hence we reject the null hypothesis
### therefore Q values of both the variants vary
